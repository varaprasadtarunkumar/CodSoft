```python
import tkinter as tk

# Function to perform calculations
def calculate():
    expression = entry.get()
    try:
        result = eval(expression)
        entry.delete(0, tk.END)
        entry.insert(tk.END, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

# Create a window
window = tk.Tk()
window.title("Calculator")

# Create an entry widget for displaying and entering calculations
entry = tk.Entry(window, width=18, justify=tk.RIGHT, font=("Arial", 14))
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

# Create buttons for numbers and operators
buttons = [
    ("7", 1, 0), ("8", 1, 1), ("9", 1, 2), ("/", 1, 3),
    ("4", 2, 0), ("5", 2, 1), ("6", 2, 2), ("*", 2, 3),
    ("1", 3, 0), ("2", 3, 1), ("3", 3, 2), ("-", 3, 3),
    ("0", 4, 0), (".", 4, 1), ("^", 4, 2), ("+", 4, 3)
]

# Create buttons and assign their respective positions and functions
for button in buttons:
    text, row, col = button
    btn = tk.Button(window, text=text, width=5, height=2, font=("Arial", 14),
                    command=lambda text=text: entry.insert(tk.END, text))
    btn.grid(row=row, column=col, padx=5, pady=5)

# Create a button for clearing the entry
clear_btn = tk.Button(window, text="C", width=5, height=2, font=("Arial", 14),
                    command=lambda: entry.delete(0, tk.END))
clear_btn.grid(row=5, column=0, padx=5, pady=5)

# Create a button for calculating the expression
equal_btn = tk.Button(window, text="=", width=5, height=2, font=("Arial", 14),
                    command=calculate)
equal_btn.grid(row=5, column=1, columnspan=2, padx=5, pady=5)

# Calculate the center position of the screen
screen_width = window.winfo_screenwidth()
screen_height = window.winfo_screenheight()
window_width = 210
window_height = 260
x = (screen_width - window_width) // 2
y = (screen_height - window_height) // 2

# Set the window's dimensions and position
window.geometry(f"{window_width}x{window_height}+{x}+{y}")

# Start the main event loop
window.mainloop()
```
