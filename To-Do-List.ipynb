```python
import tkinter as tk
from tkinter import messagebox
from datetime import datetime

class TodoListApp:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("To-Do List App")
        self.window.geometry("400x400")
        
        self.task_list = []
        
        self.create_widgets()
        
    def create_widgets(self):
        self.task_label = tk.Label(self.window, text="Task:", font=("Helvetica", 14))
        self.task_label.pack(pady=10)
        
        self.task_entry = tk.Entry(self.window, width=30, font=("Helvetica", 12))
        self.task_entry.pack()
        
        self.time_label = tk.Label(self.window, text="Time:", font=("Helvetica", 14))
        self.time_label.pack(pady=5)
        
        self.time_entry = tk.Entry(self.window, width=10, font=("Helvetica", 12))
        self.time_entry.pack()
        
        self.add_button = tk.Button(self.window, text="Add Task", command=self.add_task, bg="#4CAF50", fg="white", font=("Helvetica", 12))
        self.add_button.pack(pady=10)
        
        self.task_listbox = tk.Listbox(self.window, width=50, font=("Helvetica", 12))
        self.task_listbox.pack()
        
        self.delete_button = tk.Button(self.window, text="Delete Task", command=self.delete_task, bg="#FF0000", fg="white", font=("Helvetica", 12))
        self.delete_button.pack(pady=10)
        
        self.window.mainloop()
        
    def add_task(self):
        task = self.task_entry.get()
        time = self.time_entry.get()
        
        if task and time:
            task_with_time = f"{task} (Time: {time})"
            self.task_list.append(task_with_time)
            self.task_listbox.insert(tk.END, task_with_time)
            self.task_entry.delete(0, tk.END)
            self.time_entry.delete(0, tk.END)
        else:
            messagebox.showwarning("Empty Fields", "Please enter both task and time.")
            
    def delete_task(self):
        selected_index = self.task_listbox.curselection()
        
        if selected_index:
            self.task_listbox.delete(selected_index)
            del self.task_list[selected_index[0]]
        else:
            messagebox.showwarning("No Task Selected", "Please select a task to delete.")
        
todo_app = TodoListApp()
```
