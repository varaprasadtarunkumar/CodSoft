```python
import tkinter as tk
import random
import string

def generate_password():
    length = length_entry.get()

    try:
        length = int(length)
        if length <= 0:
            password_label.configure(text="Invalid length")
            return
    except ValueError:
        password_label.configure(text="Invalid length")
        return

    password = ''.join(random.choices(string.ascii_letters + string.digits + string.punctuation, k=length))
    password_label.configure(text=password)

# Create a window
window = tk.Tk()
window.title("Password Generator")

# Function to clear the entry field
def clear_entry():
    length_entry.delete(0, tk.END)

# Function to copy the generated password to the clipboard
def copy_password():
    password = password_label.cget("text")
    if password:
        window.clipboard_clear()
        window.clipboard_append(password)
        window.update()

# Create a label for the length entry
length_label = tk.Label(window, text="Password Length:")
length_label.pack()

# Create an entry for the length of the password
length_entry = tk.Entry(window, width=30)
length_entry.pack()

# Create a button to generate the password
generate_button = tk.Button(window, text="Generate Password", command=generate_password)
generate_button.pack()

# Create a label to display the generated password
password_label = tk.Label(window, text="", font=("Arial", 14), wraplength=250)
password_label.pack()

# Create a button to copy the generated password to clipboard
copy_button = tk.Button(window, text="Copy Password", command=copy_password)
copy_button.pack()

# Create a button to clear the length entry
clear_button = tk.Button(window, text="Clear", command=clear_entry)
clear_button.pack()

# Set the window's dimensions
window.geometry("300x300")

# Start the main event loop
window.mainloop()
#/J'QwW+1
```
